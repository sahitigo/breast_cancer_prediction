## Problem Statement 

Breast cancer is a common cancer in women, accounting for 25% of cases worldwide. The challenge is classifying tumors as malignant or benign. We'll use machine learning on the Breast Cancer Wisconsin Dataset to build accurate classification models, fine-tune hyperparameters, and compare evaluation metrics.

## Exploratory Data Analysis: 

Exploratory Data Analysis (EDA) is a vital step in understanding the distribution and characteristics of the dataset. It assists in identifying patterns, outliers, and relationships between variables. In this specific dataset, which consists of 569 observations and 32 features, conducting EDA is crucial for determining which features are relevant in predicting whether a tumor is malignant or benign.

```python
df = apply_eda(df,drop_null='no',fill_null='mean',columns_to_drop=['id'])
```

We are checking for multicollinearity by calculating VIF. Then by sorting them in descending order, removed those features which has VIF<10.Thereby retaining features, which are actually significant

   - Let's perform further steps of the EDA by importing the apply_eda function

```python
df.dtypes
```

- We are now left with 9 columns after removing multicollinear features.

```python
df.head()
```

```python
df.corr()
```

- It can be seen that the features are no more correlated.

## Model Prediction:

 **Label** : Diagnosis.

**Features** : All the columns except the Diagnosis.

```python
y = df['diagnosis_B']
X = df.drop(columns=['diagnosis_M','diagnosis_B'],axis=1)

classification_models(X, y, max_depth=2, leaf_nodes=3, n_neighbors=5)
```
